```python
import csv
import re
import os
```

```python
file_in = "prefix distribution/bgp-p_prefix-vector-6447.txt"
file_out_path = "prefix distribution/summary"
if not os.path.isdir(file_out_path):
    os.mkdir(file_out_path)
file_out = "prefix-6447-dis.csv"
file_out = file_out_path+"/"+file_out
```

```python
csvfile = open(file_out,"w",newline='')
fout = csv.writer(csvfile)
csvtitle = []
# title 為 0~32
for i in range(1,33):
    csvtitle.append(str(i))
# print(csvtitle)

fout.writerow(csvtitle)
```

```python
pattern1 = r'\d+\.+\d*'
fin = open(file_in,"r")
line = fin.readlines()


# print(re.findall(pattern1,line[1000]))
#收集五年的資料
start = len(line)-365*5
for i in line[start:]:
    temp = re.findall(pattern1,i)
    if(temp):
        fout.writerow(temp)
fin.close()
csvfile.close()
    
```

```python
import pandas as pd
data = pd.read_csv(file_out)
title = data.columns
print(len(data))
```

```python
import plotly.graph_objects as go
import numpy as np
axis = dict(showline=True, # hide axis line, grid, ticklabels and  title
            zeroline=False,
            showgrid=False,
            linecolor = "black"
            
            )

fig = go.Figure()
rowsnum = len(data)

prelen = []
distribut = []
for i in range(rowsnum):
    for j in title:
        prelen.append(j)
    for j in data.iloc[i]:
        distribut.append(j)
    

fig.add_trace(go.Scatter(
        x = prelen,
        y = distribut,
        mode='markers',
        marker_symbol=200,
        marker=dict(
            color='rgba(135, 206, 250, 0.5)',
            size=10,
            line=dict(
                color='MediumPurple',
                width=1
            )
        ),
        showlegend=False
    ))
fig.update_layout(
    plot_bgcolor='rgb(255,255,255)',
    xaxis = axis,
    yaxis = axis,
    title = "update time with one router"
    
)
fig.update_xaxes(
    title = "prefix length"
)
fig.update_yaxes(
    title = "% prefix distribution"
)
fig.show()
```
